# Lesson 7. The Simple Linear Regression &mdash; Part 2

## Example

We want to predict the price of a used Honda Accord from its mileage.

The data for 30 used Accords is in the `AccordPrice` data frame in the `Stat2Data` package.

First, let's load this data frame, like this:

```r
library(Stat2Data)
data(AccordPrice)
```

Before moving on, let's take a quick look at the data:

```r
head(AccordPrice)
```

Next, let's make a scatterplot to determine whether the data exhibits a linear trend:

```r
plot(AccordPrice$Mileage, AccordPrice$Price)
```

We fit a simple linear regression model in R like this:

```r
fit <- lm(Price ~ Mileage, data = AccordPrice)
```

We can view the fitted line on top of the scatter plot with the following code:

```r
plot(AccordPrice$Mileage, AccordPrice$Price)
abline(fit)
```

We can output the fitted model like this:

```r
summary(fit)
```

Finally, we can get more output about how well our model fits:

```r
anova(fit)
```
